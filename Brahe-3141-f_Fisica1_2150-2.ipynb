# Fluctuaciones de gravedad

se estudia la variacion de la gravedad con respecto a la media, para eso se computa F=1-g/< g > y se grafica con respecto a theta


```python
F=1-(gravity2/gravity2.mean())
scatter(theta2,F)
xlabel('Angulo polar')
ylabel('F')
title('F en funcion de theta')
show()
print gravity2.mean()
```

Se pasan los valores de F y theta a un archivo llamado fluctuaciones.dat 
el cual se pasa luego a ser ajustado por minimos cuadrados en dos programas
distintos uno para el primer modelo con coseno y otro para el segundo con
funciones cuadráticas. Posteriormente se imprimen los coeficientes de mejor 
ajuste a cada modelo y con esto se grafica el error relativo para finalmente 
concluir.

```python
f2=open("fluctuaciones.dat",'w')
for i in range (0,884):
    f2.write("%f %f \n" % (theta2[i], F[i])) 

!cc ResolverModelo1.c
!./a.out fluctuaciones.dat

au=!awk '{print $1}' ai_coef.dat
a=np.zeros(2)
for i in range(0,2):
    a[i-1]=float(au[i])
print a
modelo1=a[0]*cos(2*theta2*3.14159/180.0) + a[1]
plot(theta2,abs(modelo1-F)/F)
xlabel('angulo polar')
ylabel('Error del modelo')
```

```python
!cc ResolverModelo2.c
!./a.out fluctuaciones.dat
bu=!awk '{print $1}' bi_coef.dat
b=np.zeros(3)
for i in range(0,3):
    b[i]=float(bu[i])
print b

modelo2=b[0]+b[1]*theta2 + b[2]*theta2**2
plot(theta2,abs(modelo2-F)/F,'red')
xlabel('angulo polar')
ylabel('Error del modelo')
```

Se nota que el error en el segundo modelo diverge mas en los meridianos del planeta y por ende se considera que el primer modelo es mas acertado para la descripcion de la variacion de la gravedad con respecto al ángulo polar, entre aspectos que se deben tener en cuenta en la ejecucion de este notebook es que hay algo de inestabilidad en el calculo de minimos cuadrados en la primera parte lo que genera que ocacionalmente archivos den como resultado del ajuste como resultado nan, y la media en fluctuaciones no se pueda calcular, si esto pasa solo se ejecuta de nuevo y se arregla el problema, el error tambien se genera si %pylab inline no se ejecuta adecuadamente.
